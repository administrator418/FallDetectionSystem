```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 数据集类别
dataset_classes = {"/m/01g317":"Person" ,"/m/0dzct": "Human face"}

# 统计数据集中各类别的数量
dataset_num = {"total": 0, "Person": 0, "Human face": 0}

# 遍历数据集
dataset_category = ["train", "validation", "test"]
for i in dataset_category:
    path = f"/Users/jayden/Library/CloudStorage/OneDrive-jaydentang/Datasets/open-images-v7-face/{i}/labels/detections.csv"
    df = pd.read_csv(path)
    for j in df["LabelName"]:
        if j in dataset_classes:
            if j == "/m/01g317":
                dataset_num["Person"] += 1
                dataset_num["total"] +=1
            elif j == "/m/0dzct":
                dataset_num["Human face"] += 1
                dataset_num["total"] +=1

# 打印统计结果
print(dataset_num)
```

```python
# 画出柱状图
bars = plt.bar(dataset_num.keys(), dataset_num.values())

# 在柱状图上显示具体数值
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), va="bottom", ha="center")

# 添加标题
plt.title("Face Detection Dataset Detail")

# 显示柱状图
plt.show()
```
