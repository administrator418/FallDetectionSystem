```python
from ultralytics import YOLO

modelpath = '/kaggle/input/falldetectiondata/FallDetectionData/yolov8n_fall2.yaml'
datapath = '/kaggle/input/falldetectiondata/FallDetectionData/fall2.yaml'

# Load a model
model = YOLO(modelpath)  # build a new model from YAML
model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)
model = YOLO(modelpath).load('yolov8n.pt')  # build from YAML and transfer weights

# Train the model
results = model.train(data=datapath, epochs=100, imgsz=640)
```

```python
import os
import zipfile
import datetime

def file2zip(packagePath, zipPath):
    '''
  :param packagePath: 浠跺す璺寰
  :param zipPath: 缂╁璺寰
  :return:
  '''
    zip = zipfile.ZipFile(zipPath, 'w', zipfile.ZIP_DEFLATED)
    for path, dirNames, fileNames in os.walk(packagePath):
        fpath = path.replace(packagePath, '')
        for name in fileNames:
            fullName = os.path.join(path, name)
            name = fpath + '\\' + name
            zip.write(fullName, name)
    zip.close()


if __name__ == "__main__":
    # 浠跺す璺寰
    packagePath = '/kaggle/working/runs'
    zipPath = '/kaggle/working/runs.zip'
    if os.path.exists(zipPath):
        os.remove(zipPath)
    file2zip(packagePath, zipPath)
    print("瀹")
    print(datetime.datetime.utcnow())
```
